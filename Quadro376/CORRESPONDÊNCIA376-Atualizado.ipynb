```python
###########################################
##Criando Função Para criação dos valores##
###########################################


def valdata(date,ram):


    # Selecionando os TPMOIDs com CMPID 1001
                                                                    # Criando mascara para filtrar colunas de meu interesse
    mask_cmpid_1001 = (df['CMPID']==1001) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1001]['fESRVALORMOV']),2))
    cmpid.append(12184)
    desc_cmpid.append('Direto - Administrativo')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1002

    mask_cmpid_1002 = (df['CMPID']==1002) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1002]['fESRVALORMOV']),2))
    cmpid.append(12185)
    desc_cmpid.append('Cosseguro Aceito – Administrativo')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1003

    mask_cmpid_1003 = (df['CMPID']==1003) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1003]['fESRVALORMOV']),2))    
    cmpid.append(12186)
    desc_cmpid.append('Cosseguro Cedido – Administrativo') # Criando descrição
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1004

    mask_cmpid_1004 = (df['CMPID']==1004) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1004]['fESRVALORMOV']),2))
    cmpid.append(12208)
    desc_cmpid.append('Salvados e Ressarcidos – Administrativo')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1005

    mask_cmpid_1005 = (df['CMPID']==1005) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1005]['fESRVALORMOV']),2))
    cmpid.append(12224)
    desc_cmpid.append('Salvados e Ressarcidos ao Ressegurador - Administrativo')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1006

    mask_cmpid_1006 = (df['CMPID']==1006) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1006]['fESRVALORMOV']),2))
    cmpid.append(12188)
    desc_cmpid.append('Direto - Judicial')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1007

    mask_cmpid_1007 = (df['CMPID']==1007) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1007]['fESRVALORMOV']),2))
    cmpid.append(12189)
    desc_cmpid.append('Cosseguro Aceito – Judicial')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1008

    mask_cmpid_1008 = (df['CMPID']==1008) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1008]['fESRVALORMOV']),2))
    cmpid.append(12190)
    desc_cmpid.append('Cosseguro Cedido – Judicial')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1009

    mask_cmpid_1009 = (df['CMPID']==1009) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1009]['fESRVALORMOV']),2))
    cmpid.append(12209)
    desc_cmpid.append('Salvados e Ressarcidos – Judicial')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1010

    mask_cmpid_1010 = (df['CMPID']==1010) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)

    data.append(date)
    valor.append(round(sum(df.loc[mask_cmpid_1010]['fESRVALORMOV']),2))
    cmpid.append(12225)
    desc_cmpid.append('Salvados e Ressarcidos ao Ressegurador - Judicial')
    ramc.append(ram)


    # Selecionando os TPMOIDs com CMPID 1011

    mask_cmpid_1011 = (df['CMPID']==1011) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)
    df_cmpid_1011 = df.loc[mask_cmpid_1011]

    campo_cmpid_1011 = round(sum(df_cmpid_1011['fESRVALORMOV']),2)


    # Selecionando os TPMOIDs com CMPID 1012
 
    mask_cmpid_1012 = (df['CMPID']==1012) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)
    df_cmpid_1012 = df.loc[mask_cmpid_1012]

    campo_cmpid_1012 = round(sum(df_cmpid_1012['fESRVALORMOV']),2)


    # Selecionando os TPMOIDs com CMPID 1013

    mask_cmpid_1013 = (df['CMPID']==1013) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)
    df_cmpid_1013 = df.loc[mask_cmpid_1013]
    
    campo_cmpid_1013 = round(sum(df_cmpid_1013['fESRVALORMOV']),2)


    # Selecionando os TPMOIDs com CMPID 1014

    mask_cmpid_1014 = (df['CMPID']==1014) & ((df['TPMOID']==1)|(df['TPMOID']==2)|(df['TPMOID']==5)|(df['TPMOID']==6)) & (df['MESano'] == date) & (df['RAMCODIGO'] == ram)
    df_cmpid_1014 = df.loc[mask_cmpid_1014]
    
    campo_cmpid_1014 = round(sum(df_cmpid_1014['fESRVALORMOV']),2)


    # Criando campos com mais de um CMPID
    # Campo Recuperação de Sinistros não Pagos + já Pagos – Administrativo
    campo_cmpid_1011_1012 = round((sum(df_cmpid_1011['fESRVALORMOV'])) + (sum(df_cmpid_1012['fESRVALORMOV'])),2)

    data.append(date)
    valor.append(campo_cmpid_1011_1012)
    cmpid.append(12218)
    desc_cmpid.append('Recuperação de Sinistros não Pagos + já Pagos – Administrativo')
    ramc.append(ram)

    # Campo Recuperação de Sinistros não Pagos + já Pagos – Judial
    campo_cmpid_1013_1014 = round((sum(df_cmpid_1013['fESRVALORMOV'])) + (sum(df_cmpid_1014['fESRVALORMOV'])),2)

    data.append(date)
    valor.append(campo_cmpid_1013_1014)
    cmpid.append(12219)
    desc_cmpid.append('Recuperação de Sinistros não Pagos + já Pagos - Judicial')
    ramc.append(ram)

    # Campo cmpid_1001_2_3_6_7_8
    campo_cmpid_1001_2_3_6_7_8 = (round(sum(df.loc[mask_cmpid_1001]['fESRVALORMOV']),2)) + (round(sum(df.loc[mask_cmpid_1002]['fESRVALORMOV']),2)) - (round(sum(df.loc[mask_cmpid_1003]['fESRVALORMOV']),2)) + (round(sum(df.loc[mask_cmpid_1006]['fESRVALORMOV']),2)) + (round(sum(df.loc[mask_cmpid_1007]['fESRVALORMOV']),2)) - (round(sum(df.loc[mask_cmpid_1008]['fESRVALORMOV']),2))
    data.append(date)
    valor.append(campo_cmpid_1001_2_3_6_7_8)
    cmpid.append(12234)
    desc_cmpid.append('Despesas Financeiras - PSL')
    ramc.append(ram)

    # Campo cmpid_1011_12_13_14
    campo_cmpid_1011_12_13_14 = (campo_cmpid_1011) + (campo_cmpid_1012) + (campo_cmpid_1013) + (campo_cmpid_1014)
    data.append(date)
    valor.append(campo_cmpid_1011_12_13_14)
    cmpid.append(12235)
    desc_cmpid.append('Recuperação das Despesas Financeiras - PSL')
    ramc.append(ram)

    # Finalizando
    print('Para Data {} e Ramo {} campos analisados e dados coletados.\n'.format(date,ram))  # Gernado Print informativo
```

# Puxando Base de Dados FIP (em CSV)

```python
df_comparacao = pd.read_csv('ValoresMovRamos.csv',sep=';')  # Lendo o Arquivo CSV
```

```python
# Passando entCodigo para int e vdrValor para float e mrfMesAno para date 
df_comparacao['entCodigo'] = df_comparacao['entCodigo'].astype('int64') # Valor int entCodigo

df_comparacao['vdrValor'] = df_comparacao['vdrValor'].apply(lambda x : x.replace(",", '.')) 
df_comparacao['vdrValor'] = df_comparacao['vdrValor'].astype('float64')  # Valor float vdrValor
df_comparacao['vdrValor'] = df_comparacao['vdrValor'].apply(lambda x : round(x,2))  # Arredondando vdrValor com duas casas decimais

df_comparacao['CMPID'] = df_comparacao['CMPID'].astype('int64') # Valor int CMPID

df_comparacao['ramo_s1'] = (df_comparacao['graCodigo'].astype(str) + df_comparacao['ramCodigo'].astype(str))
df_comparacao['ramo_s1'] = df_comparacao['ramo_s1'].astype('int64')

import datetime
list_date = []   #Criando lista
for i in df_comparacao['mrfMesAno']:
    format_str = '%d/%m/%Y 00:00:00'
    datetime_obj = datetime.datetime.strptime(i, format_str).date()   # Transformando cada item para Datetime
    list_date.append(datetime_obj)      # Adicionando a lista
df_comparacao.loc[:,'DTmrfMesAno'] = list_date    # Alocando a lista no DataFrame
df_comparacao['DTmrfMesAno'] = df_comparacao['DTmrfMesAno'].astype('datetime64')
df_comparacao = df_comparacao.dropna(axis=1) # Eliminando colunas com valores Nulos
df_comparacao.head()  # Visualisando DataFrame gerado
```

# Comparando valores do QE com FIP

```python
df_merge = pd.merge(df_geral_qe, df_comparacao, on=['entCodigo','CMPID','DTmrfMesAno','ramo_s1'], how='inner',suffixes=('_qe', '_fip'))
df_merge = df_merge.drop(['mrfMesAno','ramCodigo','graCodigo'],axis=1)
```

```python
#########################
#Arquivo Modelo Empresas#
#########################

arq_codsusep = pd.read_excel('Quadro376.xlsx','Cod_SUSEP',fileEncoding = "UTF-8",encoding = 'latin_1', dtype=str) # Lendo arquivo parametro
ent = []                                          # Criando lista vazia
for i in arq_codsusep['Cod_SUSEP'].astype('int64'):
    ent.append(i)                                # Adicionando Todas as EntidadeCod em uma lista
desc = []                                        # Criando lista vazia
for i in arq_codsusep['Empresa']:
    desc.append(i)                                # Adicionando Todos os Nomes de Empresa em uma lista
    
mod = dict(zip(ent, desc))                        # Crando dicionario
#Transformalo em Data Frame
df_mod_e = pd.DataFrame(columns=['entCodigo','Empresa'])
df_mod_e.loc[:,'entCodigo'] = mod.keys()
df_mod_e.loc[:,'Empresa'] = mod.values()

##########################
#Arquivo Modelo RAMCODIGO#
##########################

arq_ramcod = pd.read_excel('Quadro376.xlsx','RAMCODIGO',fileEncoding = "UTF-8",encoding = 'latin_1', dtype=str) # Lendo arquivo parametro
ent = []                                          # Criando lista vazia
for i in arq_ramcod['ramo_s1'].astype('int64'):
    ent.append(i)                                # Adicionando Todas as EntidadeCod em uma lista
desc = []                                        # Criando lista vazia
for i in arq_ramcod['ramNome']:
    desc.append(i)                                # Adicionando Todos os Nomes de Empresa em uma lista
    
mod = dict(zip(ent, desc))                        # Crando dicionario
#Transformalo em Data Frame
df_mod_r = pd.DataFrame(columns=['ramo_s1','Desc_ramo'])
df_mod_r.loc[:,'ramo_s1'] = mod.keys()
df_mod_r.loc[:,'Desc_ramo'] = mod.values()
```

```python
# Juntando empresa com descrição a partir do arquivo modelo
df_merge = pd.merge(df_merge,df_mod_e,on='entCodigo',how='inner')
df_merge = pd.merge(df_merge,df_mod_r,on='ramo_s1',how='inner')
df_merge = df_merge[['DTmrfMesAno','vdrValor_qe','vdrValor_fip','CMPID','Desc_CMPID','entCodigo','Empresa','ramo_s1','Desc_ramo']]
df_merge
```

```python
mask_erro = (df_merge['vdrValor_qe'].astype('int64') != df_merge['vdrValor_fip'].astype('int64') )
df_erros = df_merge.loc[mask_erro]
df_erros
```

```python
# Adicionando Colunas Extras nos DataFrames
from datetime import datetime
data_presente = datetime.now()    # Obtendo data presente
nome = input("Insira seu nome: ") # Obtendo nome
df_merge.loc[:,'Criador'] = nome ### insere a coluna com o nome do criador da tabela
df_merge.loc[:,'Data do output'] = data_presente ### insere a coluna com a data de criacao da tabela
if len(df_erros) > 0:
    df_erros.loc[:,'Criador'] = nome ### insere a coluna com o nome do criador da tabela
    df_erros.loc[:,'Data do output'] = data_presente ### insere a coluna com a data de criacao da tabela
```
